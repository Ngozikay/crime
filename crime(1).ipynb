what time of the day are crimes most frequent and which week and months have the most crimes.
whats the trend of crime per year? is it increasing or decreasing?


```python

df.Start_Date_Time.dt.hour #to bring out hour from timestamp
```

```python
sns.distplot(df.Start_Date_Time.dt.hour,bins=24,kde=False, norm_hist=True)
```

#high percentage of crime occured around 12am in the morning while low percentage of crime occured between 4-7am in the morning probably beacue peopkle are awake and alert.
#next percentage of crime occured around 12pm.

```python
sns.distplot(df.Start_Date_Time.dt.dayofweek,bins=7,kde=False, norm_hist=True)
```

#Crimes are evenly distributed among the days of the week but low during weekend probably because people are at home.

```python
sns.distplot(df.Start_Date_Time.dt.year,bins=24,kde=False, norm_hist=True)
```

There is increase in the record of crimes in the year 2017 which dropped down as the year goes by and also recorded a steep fall in 2022.

QUESTION:

```python
Mondays_start_time=df.Start_Date_Time[df.Start_Date_Time.dt.dayofweek==0] 
sns.distplot(Mondays_start_time.dt.hour,bins=24,kde=False, norm_hist=True)
```
